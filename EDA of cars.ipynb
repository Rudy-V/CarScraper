```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Clean_cars_26_01_2023.csv")
```

```python
df = df[(df['year'] > 2010) & (df['milage (Thousands)'] < 100.0)]
```

```python
df.describe()
```

```python
sns.set_theme(style="ticks")
sns.pairplot(df,hue='manufacturer')
```

```python
print("Manufacturer     Market share%")
df['manufacturer'].value_counts(normalize=True)*100
```

```python
man_list = ["Ford","Renault","Volkswagen","Hyundai","Nissan","Toyota","Kia","Honda","Suzuki","Opel"]

for i in man_list:
    temp_df = df[df['manufacturer'] == str(i)]
    sns.pairplot(temp_df,hue='model', diag_kind="hist")
    plt.title(str(i))
```
